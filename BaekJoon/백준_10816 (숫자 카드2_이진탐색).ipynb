```python
from collections import Counter

n = int(input())
arr = list(map(int,input().split()))
m = int(input())
card_list = list(map(int, input().split()))

c = Counter(arr)
ans = []

for card in card_list:
    ans.append(c.get(card)) if card in c else ans.append(0)

print(*ans)
```
