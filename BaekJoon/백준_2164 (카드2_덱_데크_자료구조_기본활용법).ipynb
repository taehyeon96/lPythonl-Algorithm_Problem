```python
'''  2164 -   =>         '''

from collections import deque

N = int(input())
deque = deque([i for i in range(1, N+1)])  #  

while len(deque) > 1:
    deque.popleft()
    deque.rotate(-1)      #     
    
print(deque[0])
```

```python
from collections import deque

n = int(input())
que = deque([i for i in range(1,n+1)])


while que:
    if len(que) > 1:
        que.popleft()
    else:
        print(que[0])
        break        
    que.append(que.popleft())
```
