```python
'''  15649 -  ( ) '''
from itertools import permutations     # iter tools  (<-->  combinations)

n, m = map(int, input().split())

#      , ()    str()  
result = list(permutations([str(i) for i in range(1, n+1)], m)) 

for i in result:
    print(" ".join(i))

#                   = ,    = nCm , m
```
