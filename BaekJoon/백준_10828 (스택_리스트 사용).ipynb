```python
'''   -   '''
#   1. input sys 
import sys

result = []
for _ in range(int(sys.stdin.readline())):
    cmd = sys.stdin.readline()
    
    if cmd[:3] == "pus": result.append(cmd.split()[1])
    elif cmd == "pop" and result:
        print(result.pop())
    elif cmd == "pop" and not result:
        print(-1)
    elif cmd == "size": print(len(result))
    elif cmd == "empty" and result: print(0)
    elif cmd == "empty" and not result: print(1)
    elif cmd == "top" and not result: print(-1)
    elif cmd == "top" and result: print(result[-1])
    
```

```python
'''  '''
import sys

#  X   .
def push(x):
    stack.append(x)

#      ,   . 
#       -1 .
def pop():
    if(not stack):
        return -1
    else:
        return stack.pop()

#     .
def size():
    return len(stack)

#   1,  0 .
def empty():
    return 0 if stack else 1

#      . 
#       -1 .
def top():
    return stack[-1] if stack else -1

N = int(sys.stdin.readline().rstrip())
stack = []

for _ in range(N):
    input_split = sys.stdin.readline().rstrip().split()

    order = input_split[0]

    if order == "push":
        push(input_split[1])
    elif order == "pop":
        print(pop())
    elif order == "size":
        print(size())
    elif order == "empty":
        print(empty())
    elif order == "top":
        print(top())
```
