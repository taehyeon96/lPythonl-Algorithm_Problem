```python
'''   (   ) '''
n, b = map(int, input().split())

result = []
def change(n, b):
    while True:
        result.append(n % b)
        if n // b < b:
            result.append(n//b)
            return result
        else:
            return change(n//b, b)
        

print(change(n,b))
```

```python
'''   (   ) '''
n, b = map(int, input().split())

def base10(n,b):
    if n==0: return [0]
    
    result = []
    while n>0:
        n, m = divmod(n, b)   # () n=,  m=
        result.insert(0, m)   #    (m)
    
    return result

print(base10(n,b))
```

```python
''' ()   '''
def encode(digits, digit_map):    # digits=change() , digit_map= 
    encoding = ''    
    for d in digits:
        encoding += digit_map[d]
    return encoding

print(encode([15,15], '0123456789ABCDEF'))
```

```python
'''    '''
n, b = map(int, input().split())

def total_base10(n,b):

    digit_map = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'   #  
    
    if b < 2 or b > 36:  # 36 !
        raise ValueError("invalid base: 2<=b<=36")
    
    digits = base10(n, b)
    encoding = encode(digits, digit_map)
    
    return encoding

print(total_base10(n,b))
```

```python
'''  '''
n, b = map(int, input().split())

def total(n,b):    
    result = []
    
    while n > 0:
        n, m = divmod(n, b)   # () n=,  m=
        result.insert(0, m)   #    (m)
    
    digit_map = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'   #  
    
    encoding = ''    
    for d in result:
        encoding += digit_map[d]
        
    return encoding

print(total(n,b))
```
