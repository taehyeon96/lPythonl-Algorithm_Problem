```python
'''   '''

from collections import Counter

arr = input()

arr = arr.upper()                    # ()     .upper() 

result = Counter(arr).most_common()  #   Counter().most_common() 

if len(result) == 1:                 #   
    print(result[0][0])
elif result[0][1] == result[1][1]:   #    = 2   =    
    print('?')
else:                                 #  ,     
    print(result[0][0])
```
