```python
'''   - )     =>   sort() '''
from collections import deque

#---------------------------------------------------------------------#
def DFS(check, graph, v):
    check[v] = True               #  
    print(v, end = ' ')
                                   # ()   -> loop 
    for i in graph[v]:            # () v  
        if check[i] == False:     # () v   
            DFS(check, graph, i)
#---------------------------------------------------------------------#
            
def BFS(check, graph, start):
    que = deque([start])            # () BFS  
    check[start] = True            
    
    while que:                     # ()    -> loop 
        v = que.popleft()
        print(v, end = ' ')
        
        for i in graph[v]:
            if not check[i]:
                que.append(i)        # ()      
                check[i] = True     # ()   
#---------------------------------------------------------------------#

n, m, v = map(int, input().split())  
arr = [list(map(int, input().split())) for _ in range(m)]  

check = [False] * (n+1)             # () False  

# Create Graph ( )
graph = [[] for _ in range(n+1)]
for i in range(len(arr)):
    graph[arr[i][0]].append(arr[i][1])
    graph[arr[i][1]].append(arr[i][0])

#  1260 
for i in range(len(graph)):
    graph[i].sort()
#---------------------------------------------------------------------#

DFS(check, graph, v)
print()
check = [False] * (n+1)            # ()   ......
BFS(check, graph, v)
```
