## 1.    (Memorization)

```python
'''   -   '''

result = []
arr = [0]*100   # 10  1000 , 100 ( )

for i in range(10):
    result.append(int(input()))
    arr[result[i]//10] += 1      # 1000  arr[result[i]]   
    
print(sum(result)//10)
print(arr.index(max(arr))*10)    #  
```

## 2. collections Counter( )   ( )

```python
'''   -   '''
from collections import Counter

result = [int(input()) for _ in range(10)]   # ()

print(sum(result)//10)

a = Counter(result).most_common()
print(a[0][0])                       # a   
```
