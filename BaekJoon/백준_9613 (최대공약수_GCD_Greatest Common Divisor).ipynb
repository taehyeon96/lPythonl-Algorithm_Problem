```python
'''     '''
#  DFS & BFS 
def gcd(a, b):
    if a%b == 0:
        return b
    else:
        return gcd(b, a%b)
    
t = int(input())
result = []
for _ in range(t):
    total=0
    n, *arr = map(int,input().split())
    
    for j in range(n):
        for k in range(j+1, n):
            total += gcd(arr[j], arr[k])
    result.append(total)
    
#print(*result)
print("\n".join(map(str,result)))
```
