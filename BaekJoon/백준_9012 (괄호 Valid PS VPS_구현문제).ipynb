```python
'''   -   '''
def is_vpn(arr):
    num = 0
    for i in range(len(arr)):
        if arr[i] == '(': num += 1
        else: 
            if num == 0: 
                num += 1    # NO   +1
                break
            else: num -= 1
    return num  
    
for _ in range(int(input())):
    print("YES") if is_vpn(input()) == 0 else print("NO")
```

```python
'''    -  (  ) '''
from sys import stdin

n = int(input())
for _ in range(n):
    str_ = stdin.readline().strip()
    stack = 0
    for chr_ in str_:
        if chr_ == '(':
            stack += 1
        else:
            stack -= 1
            if stack < 0:    #   
                break
    if stack == 0:
        print('YES')
    else:
        print('NO')
```
