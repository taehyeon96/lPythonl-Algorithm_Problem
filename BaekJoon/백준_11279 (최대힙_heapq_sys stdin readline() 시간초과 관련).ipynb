```python
'''  11279 -   - jupyter input()   '''

import heapq

heap = []

for i in range(int(input())):
    x = int(input()) * -1               # () (=)
    
    if x == 0 and len(heap) == 0:      # is empty
        print(0)
    elif x == 0:
        print(heapq.heappop(heap) * -1)  # () (=)
    else:
        heapq.heappush(heap, x)
```

```python
'''  11279 -   -    '''

import sys, heapq  
heap = []  
 
for i in range(int(sys.stdin.readline())):  
    x = int(sys.stdin.readline()) * -1  
    if x == 0:  
        if len(heap) == 0:  
            print(0)  
        else:  
            print(heapq.heappop(heap) * -1)  
    else:  
        heapq.heappush(heap, x)
```
