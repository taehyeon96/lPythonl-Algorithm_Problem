```python
'''   '''
result = []

for _ in range(4):
    yut = list(map(int,input().split()))    
    result.append({1:'A', 2:'B', 3:'C', 4:'D', 0:'E'}.get(yut.count(0), "default")) # () 0:E "" 

print("\n".join(result))
```
