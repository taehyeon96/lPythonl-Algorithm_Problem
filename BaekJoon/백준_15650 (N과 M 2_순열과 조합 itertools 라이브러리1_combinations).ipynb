```python
'''  15650 -  ( ) '''
from itertools import combinations    # iter tools (s) (<-->  permutations)

n, m = map(int, input().split())

#      , ()    str()  
result = list(combinations([str(i) for i in range(1, n+1)], m)) 

for i in result:
    print(" ".join(i))

#                   = ,    = nCm , m
```
