```python
'''   1 '''
result = []
a = 0
for _ in range(7):
    num = int(input())
    if num % 2 != 0: result.append(num)
    else: a += 1

print(-1) if a == 7 else print(sum(result), '\n', min(result), sep='')
```

```python
'''   2 '''
lst = [int(input()) for _ in range(7)]
result = [i for i in lst if i % 2 != 0]

if len(result) == 0: print("-1")
else: print(sum(result), "\n", min(result), sep='')
```
