```python
'''   (for_ in range(t)     ) '''

t = int(input())

for _ in range(t):                   #    
    a = list(bin(int(input())))[2:]
    a.reverse()
    
    for i in range(len(a)):
        if a[i] == '1': 
            print(i, end=' ')
    
```

```python
'''   ->     ,     '''
T = int(input())

for _ in range(T):
    n = bin(int(input()))[2:]

    for idx, val in enumerate(n[::-1]):   # enumerate()    ()
        if val == '1':
            print(idx, end=' ')
```
