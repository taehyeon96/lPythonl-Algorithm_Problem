```python
'''  2588   '''
a = int(input())
b = list(map(int,input()))

j = 0
result = 0

for i in range(len(b)-1,-1, -1):    #    b.reverse() 
    print(a * b[i])
    result = result + ((a * b[i]) * pow(10, j))
    j = j + 1

print(result)
```

```python
'''  2588 reverse() '''
a = int(input())
b = list(map(int,input()))

b.reverse()

result = 0

for i in range(len(b)):
    print(a * b[i])
    result = result + ((a * b[i]) * pow(10, i))

print(result)
```
