```python
''' 1 '''
def dfs(a, idx, visit, suma, target):
    visit[idx] = True
    
    print(idx)
    print(visit, suma, len(visit))
    if (idx+1) == len(visit): return 1
    
    if not visit[idx+1]:
        suma += a[0]
        dfs(a, idx+1, visit, suma, target)
    else:
        if suma == target:        return 1
        else: return 0
    return 0

def solution(numbers, target):
    answer = 0
    
    arr = [[i, i*(-1)] for i in numbers]
    visit = [False] * len(numbers)
    
    for idx, a in enumerate(arr[:-1]):
        suma = arr[-1][0]
        if not visit[idx]:
            answer += dfs(a, idx, visit, suma, target)
        print(answer)
        suma = arr[-1][1]
        if not visit[idx]:
            answer += dfs(a, idx, visit, suma, target)
        print(answer)
    
    print(arr)
    
    
    return answer
```

```python
'''   - DFS(Stack ) - [1,-1] -1  '''
def solution(numbers, target):
    answer = 0    
    n = len(numbers)
    
    # 0    [4 -4]   stack -4  
    que = [[numbers[0], 0], [-1 * numbers[0], 0]]  
    
    while que:
        print(que)
        temp, idx = que.pop()
        idx += 1
        
        if idx < n:
            que.append([temp + numbers[idx], idx])
            que.append([temp - numbers[idx], idx])
        else: # (idx == n  )
            if temp == target:
                answer += 1
                print('---------------')
    
    return answer
```

```python
'''   - BFS -   stack  count '''
def solution(numbers, target):
    sup= [0]
    print(sup)
    for i in numbers:
        sub_node = []
        for j in sup : 
            sub_node.append(j+i)
            sub_node.append(j-i)
        sup = sub_node
        print(sub_node)
        
    return sup.count(target)
```
