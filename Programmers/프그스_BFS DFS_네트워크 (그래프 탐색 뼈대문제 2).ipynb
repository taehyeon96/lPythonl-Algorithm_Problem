```python
'''  (...) '''

def solution(n, computers):
    answer = 0
    visited = [False] * n
    
    # () (main) DFS  
    for node_idx in range(n):                   # node_idx =    
        if visited[node_idx] == False:          #        
            dfs(n, computers, visited, node_idx) #     
            answer += 1                          # DFS     1
    return answer
    

#    dfs (  )
def dfs(n, computers, visited, node_idx):       # node_idx =   
    visited[node_idx] = True
    
    for other_idx in range(n):
        #    , 1 (=    )
        if other_idx != node_idx and computers[node_idx][other_idx] == 1:
            if visited[other_idx] == False:
                dfs(n, computers, visited, other_idx) #   

print(solution(3, [[1, 1, 0], [1, 1, 0], [0, 0, 1]]), end="\n\n")
print(solution(3, [[1, 1, 0], [1, 1, 1], [0, 1, 1]]))
```
