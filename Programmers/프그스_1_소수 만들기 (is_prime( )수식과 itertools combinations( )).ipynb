```python
'''   (   ) '''
from itertools import combinations       # 

def is_prime_number(x):
    for i in range(2, x):
        if x % i == 0:
            return False
    return True

def solution(nums):
    answer = 0
    a = list(combinations(nums, 3))   # nC3 
    
    for i in range(len(a)):
        if is_prime_number(sum(a[i])) == True:
            answer += 1

    return answer
```

```python
'''   -  '''
from itertools import combinations       # 

def solution(nums):
    answer = 0
    
    for comb in combinations(nums, 3):     # nC3 
        for j in range(2, sum(comb)):      # 
            if sum(comb) % j == 0: break
        else: answer += 1                  # for-else 

    return answer
```

```python
solution([1,2,3,4])
```
